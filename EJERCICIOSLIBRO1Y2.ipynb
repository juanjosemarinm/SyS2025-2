```python
#Agregar otro escalón al principio de la señal escalera modificando el ejemplo anterior del cuaderno 3.
import sympy as sym

# Se define la variable simbólica t
t = sym.symbols('t')

# Se define el pulso rectangular básico rect(t) de ancho 1, centrado en t=0
rect = sym.Heaviside(t + 0.5) - sym.Heaviside(t - 0.5)

# Se crea la señal escalera x(t)
# 1. NUEVO ESCALÓN: Altura 4/3, centrado en t = -1/2 (va de -1 a 0)
# 2. El resto de los escalones originales
x = (4/3 * rect.subs(t, t + 1/2) +   # Nuevo escalón al principio
     1 * rect.subs(t, t - 1/2) +     # Escalón original (altura 1)
     2/3 * rect.subs(t, t - 3/2) +
     1/3 * rect.subs(t, t - 5/2))

# Se grafica la nueva señal x(t)
# Se ajusta 'ylim' a 1.5 para que el nuevo escalón más alto sea visible
sym.plot(x, (t, -2, 5), ylim=[-0.2, 1.5], ylabel='$x(t)$', title='Señal Escalera Modificada');
```
